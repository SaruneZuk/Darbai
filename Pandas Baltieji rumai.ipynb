```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
duomenys = pd.read_csv('wh_staff_dataset.csv')
duomenys.head(10)
```

```python
duomenys.info()
```

```python
duomenys.groupby('year').count()['gender']
```

```python
data_frame = duomenys.groupby(['year', 'gender'])['gender'].count()
data_frame
```

```python
data_frame = duomenys.groupby(['year', 'gender'])['gender'].count().unstack()
data_frame
data_frame.plot(color={'Female': 'magenta', 'Male': 'mediumblue'})
plt.show()

```

```python
data_frame = duomenys.groupby(['year', 'gender'])['gender'].count().unstack()
data_frame
data_frame.plot.bar(subplots=True, color={'Female': 'magenta', 'Male': 'mediumblue'})
plt.show()
```

```python
data_frame = duomenys.groupby(['year', 'gender']).size().unstack()
data_frame
data_frame.plot()
plt.show()
```
